```python
from plotbot import *
```

```python
# Cell 1: Create custom variable and plot it
trange = ['2021-01-17/12:00:00', '2021-01-17/18:00:00']

my_var = custom_variable('my_var', lambda: proton.density * 2)
my_var.color = 'purple'
my_var.y_label = 'Density × 2'

plotbot(trange, my_var, 1)
```

```python
# Cell 2: Change y_limit and plot again
my_var.y_limit = [0, 3000]

plotbot(trange, my_var, 1)
```
